```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Define the color map
COLOR_MAP = {
    0: '#000000',  # Black
    1: '#0074D9',  # Blue
    2: '#FF4136',  # Red
    3: '#2ECC40',  # Green
    4: '#FFDC00',  # Yellow
    5: '#B10DC9',  # Purple
    6: '#FF851B',  # Orange
    7: '#7FDBFF',  # Light Blue
    8: '#01FF70',  # Light Green
    9: '#F012BE'   # Pink
}

# Function to display a grid
def display_grid(grid, title):
    grid_array = np.array(grid)
    rows, cols = grid_array.shape
    
    # Create an image with color-mapped cells
    plt.figure(figsize=(cols / 2, rows / 2))
    for r in range(rows):
        for c in range(cols):
            color = COLOR_MAP.get(grid_array[r, c], '#FFFFFF')  # Default to white if not in the map
            plt.gca().add_patch(plt.Rectangle((c, rows - 1 - r), 1, 1, color=color))
    
    plt.xlim(0, cols)
    plt.ylim(0, rows)
    plt.title(title)
    plt.axis('off')
    plt.gca().set_aspect('equal')
    plt.show()

# Load the JSON file
# with open('f8b3ba0a.json', 'r') as file:
# with open('e76a88a6.json', 'r') as file:
with open('ef135b50.json', 'r') as file:
    data = json.load(file)

# Display each input with its corresponding output
for i, item in enumerate(data['train']):
    input_grid = item['input']
    output_grid = item['output']

    print(f"Example {i + 1}")
    print("Input:")
    display_grid(input_grid, f"Input Grid {i + 1}")

    print("Output:")
    display_grid(output_grid, f"Output Grid {i + 1}")
```
